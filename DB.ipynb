```python
# disponibilidade de turmas, salas, professores para cada curso
create('''create view disponibilidade_professor as
select professor.cod As cod_departamento,professor.nome as nome_prof, disciplina.nome as nome_disciplina,cod_sala, horas_disponivel, disciplina_ofertada, professor.cod_turma as prof_turma,
carga_horaria_max as horas_max, dias_nao_disponiveis,sala.horario_permitido,creditos_disciplina
from professor 
inner join  disciplina on (professor.disciplina_ofertada[1]= disciplina.nome or professor.disciplina_ofertada[2]= disciplina.nome)
inner join	sala on (disciplina.cod_sala= sala.cod);''')
## salas 3 e 4 estão disponíveis ainda( nessa tabela você pode ver a relação de que estas salas estão desocupadas para possíveis realocações)


# Salas disponiveis para os semestres, turmas e suas capacidades, tipo da sala, horario de encaixa por meio do codigo de horário
create('''CREATE VIEW salas_disponiveis as
select turma.cod,curso.nome as nome_curso, cod_turma, semestre,qtde_pessoas as qtde_pessoas_semestre,capacidade as capacidade_sala, sala.cod as cod_sala , sala.tipo, sala.cod_professor from curso
inner JOIN turma on (curso.nome = turma.nome)
inner join sala on (sala.capacidade >= turma.qtde_pessoas);''')
```

```python
# 1) gerando sequência de dias de aulas

def executa_disponibilidade(df):
        dias = {1:'segunda',2:'terça',3:'quarta',4:'quinta',5:'sexta'}
        sequencias_dias= ['135','24','13','35'] # cada sequência corresponde a um possivel encaixa de horários
        sequencia_aula = []

        def checa_opcoes_dias_aula():
          for sequencia in sequencias_dias: 
              seq = []
              for elementos in sequencia:
                  seq.append(dias[int(elementos)])
              sequencia_aula.append(seq)
          return sequencia_aula


        aulas =  checa_opcoes_dias_aula()
        print("######################### Dias disponiveis ####################\n")
        print(aulas)
        print("\n###############################################################")


        # 2) Calculo de encaixe de horas


        # calculando quantidade de horas
        def calcula_qtd_dias(df):
          # Fazendo dicionário para retornar disciplinas horas por semana e o nome das disciplinas 
          disciplinas_aulas_semana ={}
          for index in range(len(df['creditos_disciplina'])):
            #calculando horas por semana de cada disciplina
            if(df.loc[index,'creditos_disciplina']==96 or df.loc[index,'creditos_disciplina']==64 or df.loc[index,'creditos_disciplina']== 32):
                    disciplinas_aulas_semana[df.loc[index,'nome_disciplina']] = df.loc[index,'creditos_disciplina']/32
          return disciplinas_aulas_semana
        horas_aula_semana = calcula_qtd_dias(df)

        print("\n\n--------> Quantidade de aulas por semana para cada disciplina  <----------\n")
        print("----------->", horas_aula_semana )
        print("\n###############################################################")


        # 3) Dias disponíveis do professor
        def checa_dias_professor(df):
          salva_dias = [] # dias que o professor pode dar aula
          dias_nao_disponiveis = df.loc[:,'dias_nao_disponiveis'].unique() # dias que o professor não pode dar aula
          qtd_aulas_max = df.loc[0,'horas_max']/2 # aulas por semana que o professor pode dar
          for seq_dias in aulas:
            for dias_aulas in seq_dias:
              if seq_dias not in dias_nao_disponiveis:
                salva_dias.append(seq_dias)
                break # para nao repetir mais de uma vez o vetor;
          return salva_dias
        print("\n\n-----------> Dias disponiveis para o professor <-------------\n")
        dias_de_aula = checa_dias_professor(df)
        print(dias_de_aula)


        # Retorna horario formatado
        def horario_formatado(cod_horario):
            if cod_horario ==810:
              return '8hrs-10hrs'
            elif cod_horario ==1012:
              return '10hrs-12hrs'
            elif cod_horario ==1416:
              return '14hrs-16hrs'
            elif cod_horario ==1618:   
              return '16hrs-18hrs'
            elif cod_horario ==1820:
              return '18hrs-20hrs'
            else:
              return '20hrs-22hrs'

        # Função para formatar palavras de dias dentro dos dias disponiveis
        def junta_palavras(vetor_string):
            texto = ''
            for elemento in vetor_string:
              texto += elemento +'-'
            return texto[0:-1]
            
        # Vê se o professor tem os requisitos de horas completos para saber se há encaixe correto do número de horas para ele
        def calcula_horas_restantes(df):
          possibilidades_de_alocacao = []
          soma_horas = 0 
          for valor in horas_aula_semana.values():
              soma_horas += valor*2;
          horas_restantes = df.loc[0,'horas_max'] - soma_horas
          if horas_restantes > 0:
            print("Ainda há:", int(horas_restantes),"horas disponíveis.")
            print("Status: Ótimo")
          elif horas_restantes == 0:
            print("Não há mais horas para o professor, mas houve encaixe correto de disciplinas.")
            print("Status: Ok")
          else:
            print("A carga de horas excedeu o que era esperado do professor.")
            print("Status: Ruim")
          
          
        #Parte de imprimir as possibilidades de encaixe de horários
          print("\n\nEsse são os horários disponíveis para o professor: ")
          # Percorrendo chaves
          for key in horas_aula_semana.keys():
            # Retornando índice das chaves
            indice = list(df['nome_disciplina']).index(key)
            for aulas in dias_de_aula:
              if len(aulas) == horas_aula_semana[key]:
                dias = junta_palavras(aulas) # retorna dias de aula formatado
                horario = horario_formatado(df.loc[indice,'horario_permitido'])
                print("<",df.loc[0,"nome_prof"],",",df.loc[indice,"nome_disciplina"],",",df.loc[0,'cod_professor'],",","Semestre "+ str(df.loc[indice,"semestre"]),",",dias,",",horario,",",df.loc[indice,'tipo'],">")

        print("\n###############################################################")
        print("\n----------------> Horas de encaixe para o professor <--------------")
        horas_restantes = calcula_horas_restantes(df)
        horas_restantes

        print("\nObs: Note que isso são sugestões de possibilidade de salas para diferentes horários, pode-se encaixar um ou outro a depender da disponibilidade dos")
```
